## Combine all CSVs

```python
df = pd.concat(
    map(pd.read_csv, ['March 30.csv', 'March 31-May 27.csv', 'May 28-July 15.csv', "July 16-Sep 2.csv", "Sep 3-Oct 18.csv", 'Oct 19-Nov 29.csv', 'Nov 30-Jan 31.csv']), ignore_index=True)
df
```

## Filter for committee

For both Trump and Biden. Primary and affiliated committees can be found at the [Federal Election Commission](https://www.fec.gov/data/candidates/president/?election_year=2024&cycle=2024&election_full=true) website. Click the candidate name and on the next page, see **FILINGS** tab.

```python
#both Biden and Trump
committees = ['C00703975', 'C00744946', 'C00838912', 'C00849281', 'C00828541', 'C00770941', 'C00855114']
coms = df[df['filer_id'].isin(committees)].reset_index(drop=True)
coms
```

## Filter by Biden

```python
committees = ['C00703975', 'C00744946', 'C00838912', 'C00849281']
biden = df[df['filer_id'].isin(committees)].reset_index(drop=True)
biden
```

## Biden contributions

Clean and filter Biden contributions. Let's start with changing the dates to proper date format.

```python
biden.transaction_date = biden.transaction_date.astype(str)
biden
```

```python
biden.transaction_date = biden.transaction_date.str.zfill(10)
biden
```

```python
biden.transaction_date = biden.transaction_date.str.replace(r'[.]\d', '', regex=True)
biden
```

```python
biden.transaction_date = pd.to_datetime(biden.transaction_date, format= '%m%d%Y')
biden
```

## Filter by Trump

```python
committees = ['C00828541', 'C00770941', 'C00855114']
trump = df[df['filer_id'].isin(committees)].reset_index(drop=True)
trump
```

## Trump contributions

Clean and filter Trump contributions. Let's start with changing the dates to proper date format.

```python
#same code as Biden
trump.transaction_date = trump.transaction_date.astype(str)
trump.transaction_date = trump.transaction_date.str.zfill(10)
trump.transaction_date = trump.transaction_date.str.replace(r'[.]\d', '', regex=True)
trump.transaction_date = pd.to_datetime(trump.transaction_date, format= '%m%d%Y')
trump
```

## Save filtered dataset

This is optional and will be used for separate notebooks for each candidate.

```python
biden.to_csv('filtered/biden individual contributions-Jan2024.csv', index=False)
trump.to_csv('filtered/trump individual contributions-Jan2024.csv', index=False)
```
